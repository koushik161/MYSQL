```python
%%sql
##WINDOW FUNCTION
CREATE TABLE Sales (
    TransactionID INT,
    Store VARCHAR(50),
    SalesAmount DECIMAL(10, 2)
);

INSERT INTO Sales (TransactionID, Store, SalesAmount)
VALUES
    (1, 'A', 100.00),
    (2, 'A', 200.00),
    (3, 'A', 150.00),
    (4, 'B', 250.00),
    (5, 'B', 300.00);
```

```python
%%sql
SELECT TransactionID, Store, SalesAmount,
       SUM(SalesAmount) OVER ( PARTITION BY Store order by TransactionID asc  ) AS TotalSales
FROM Sales;
```

```python
%%sql
use world;
CREATE TABLE Employees (
    EmployeeID INT,
    Name VARCHAR(100),
    Department VARCHAR(50),
    HireDate DATE
);


INSERT INTO Employees (EmployeeID, Name, Department, HireDate)
VALUES
    (1, 'Alice', 'HR', '2020-05-01'),
    (1, 'Alice', 'HR', '2022-06-15'),
    (2, 'Bob', 'IT', '2021-07-10'),
    (3, 'Charlie', 'Finance', '2020-09-30'),
    (3, 'Charlie', 'Finance', '2022-05-22');
```

```python
%%sql
##ROW_NUMBER
select * from Employees
```

```python
%%sql
#NOW WE ARE GOING TO GET UNIQUE NAME
with cte as  (
select EmployeeID,Name,Department,HireDate,
ROW_NUMBER() OVER (PARTITION BY EmployeeID ORDER BY HireDate asc) as rownum
from Employees) 

select EmployeeID,Name,Department,HireDate from cte  where rownum=1
```
