```python
%%sql
use sys;
###RANK and DENSE RANK
CREATE TABLE Students (
    StudentID INT,
    StudentName VARCHAR(100),
    ExamScore INT
);

INSERT INTO Students (StudentID, StudentName, ExamScore)
VALUES
    (1, 'Alice', 95),
    (2, 'Bob', 90),
    (3, 'Charlie', 95),
    (4, 'David', 85),
    (5, 'Eva', 90);
```

```python
%%sql
select * from Students
```

```python
%%sql
###RANK
select StudentID,StudentName,ExamScore,
rank() over (order by ExamScore desc) as Student_Rank
from Students;
```

```python
%%sql
###DENSE_RANK
select StudentID,StudentName,ExamScore,
dense_rank() over (order by ExamScore desc) as Student_Rank
from Students;
```
